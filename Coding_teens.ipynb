### Overwritting Variables

In Python, overwriting a variable means changing the value you already gave it. Imagine you have a box labeled `x`, and at first, you put the number `10` inside. Later, you decide to change what’s in the box and put `25` in there instead. Now, `x` only holds `25`, and the `10` is gone.

### Example:
```python
x = 10      # First, we put 10 in x
print(x)    # It prints: 10

x = 25      # Now we put 25 in x, so it forgets about 10
print(x)    # It prints: 25
```

This is useful if you want to keep updating your answer or keep track of changes without using a lot of different names. Just remember, once you change it, the old value is gone!

````python
# potentially useful things

# dictionary holding the planets' distance from the sun in astronomical units (AU)
planet_distances = {"mercury":0.39, "venus":0.72, "earth":1.00, "mars":1.52, "jupiter":5.2, "saturn":9.5, "uranus":19.2, "neptune":30.1}
scale_factor = 9

# this loop will draw a circle with a radius of `R' if a turtle is initialized
# you can try to come up with this function yourself if you would like to (just don't look at it!)
clearscreen()
initializeTurtle()
R = 60
c = 2 * math.pi * R
step = c/360
penup()
goto(0, R)
pendown()
for i in range(360):
  forward(step)
  right(1)
````

# Extra challenge: Can you figure out how to place the planets in their correct place based on a time?
If you put an number of years since year 0 where we will assume every planet was perfectly aligned. Can you figure out where the planets would all be and place them on the orbits?

````python
def get_angle(planet, years):
  r = planet_distances[planet]  # in AU

  # use Kepler's law to get the planet's period
  period = r**(3/2) # in years

  frac_of_period =  (years / period) % 1

  angle = 360 * frac_of_period  # angle in degrees
  return angle
````

````python
clearscreen()
years = 123789
goto(0,0)
dot(4, 'yellow')
for planet in planet_distances.keys():

  r = scale_factor * planet_distances[planet]
  c = 2 * math.pi * r
  step = c/360

  # penup()
  # goto(0, r)
  # down()
  # f=1
  # for i in range(360//f):
  #   forward(step*f)
  #   right(1*f)

  penup()
  goto(0,r)
  pendown()
  circle(-r)

  planet_angle = get_angle(planet, years)
  x0 = r * math.sin(math.radians(planet_angle))
  y0 = r * math.cos(math.radians(planet_angle))
  penup()
  goto(x0, y0)
  pendown()
  dot(3)
hideturtle()
````
